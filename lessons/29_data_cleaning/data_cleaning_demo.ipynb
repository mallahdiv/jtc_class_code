#### Filtering rows based on the DATA in the dataframe

we only want rows where reivews are longer than 50 words


involves LOGIC -- we only want rows where `CONDITION` is true

```python
# a column + a logical condition
reviews['reviewWords'] > 50
```

```python
# dataframename[dataframename['colummnd'] > value]
# dataframename[logical statement applied to 1 column]
reviews_over_50_words = reviews[reviews['reviewWords'] > 50]
```

```python
reviews_over_50_words.head()
```

# Finding/dealing with missing data

`.isnull()` -- a method for BOTH dataframes, and series

returns `False` if there is data, and `True` if the data is missing (cell is blank)

```python
reviews.isnull()
```

# How many missing values in each column?

```python
reviews.isnull().sum()
```

```python
# show the rows where reviewerName is missing
reviews[reviews['reviewerName'].isnull()]
```

```python
# show the rows where reviewerName is NOT missing
# TAKE OUT those rows, and save into a new dataframe called reviews_clean
reviews_clean = reviews[-reviews['reviewerName'].isnull()]
```

```python
reviews_clean
```

```python
reviews_clean.isnull().sum()
```

```python
reviews_clean = reviews_clean[-reviews_clean['overall'].isnull()]
```

```python
reviews_clean.isnull().sum()
```

# Save to csv file

```python
reviews_clean.to_csv('../../datasets/reviews_clean.csv', index = False)
```
