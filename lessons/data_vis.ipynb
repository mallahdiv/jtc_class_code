# Import packages

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Justice-Through-Code/fall_2020/master/datasets/census.csv')
```

```python
df.head()
```

```python
df.info()
```

# Histogram

One continuous variable

```python
sns.distplot(df['median_age'])
```

```python
sns.distplot(df['avg_household_size'])
```

# Distributions for categorial variables

`catplot()`

```python
df['population_group'].value_counts()
```

```python
sns.catplot(data=df, x = 'population_group', kind = 'count')
```

# Relationship between 2 variables (2 columns)

Continuous variables -- scatter plot

```python
sns.scatterplot(data = df, x = 'median_household_income', y = 'unemployment_pct')
```

### Adding line of best fit -- 'linear relationship' between 2 continuous variables

```python
sns.lmplot(data = df, x = 'median_household_income', y = 'unemployment_pct')
```

```python
sns.lmplot(data = df, x = 'median_age', y = 'median_household_income')
```

### 2 continuous variable plus a categorical one

```python
sns.lmplot(data = df, x = 'median_household_income', y = 'unemployment_pct', hue = 'population_group')
```

# Compare categories

### Bar graph

```python
sns.barplot(data = df, x = 'population_group', y = 'median_household_income')
```

```python
sns.boxplot(data = df, x = 'population_group', y = 'median_household_income')
```

```python
sns.swarmplot(data = df, x = 'population_group', y = 'median_household_income')
```

# Changing the figure size

```python
sns.set(rc={'figure.figsize':(6, 10)})
```

```python
sns.swarmplot(data = df, x = 'population_group', y = 'median_household_income')
```

# Labels

```python
sns.set(rc={'figure.figsize':(6, 6)})
g = sns.barplot(data = df, x = 'population_group', y = 'median_household_income')
g.set(xlabel = 'County Size',  ylabel = 'Median Household Income ($)', 
      title = 'Median income in small vs. large counties')
```

# plot style

```python
sns.set(style = 'ticks', font = 'Arial', palette = 'Pastel1')
```
